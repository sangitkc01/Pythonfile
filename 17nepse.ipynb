```python
from bs4 import BeautifulSoup
import requests
import csv

y=[]
n=int(input('enter number = '))
for i in range(1,n+1):
    url=BeautifulSoup(f'http://www.nepalstock.com/main/todays_price/index/{i}/','html.parser')
    response=requests.get(url)
    data=response.text
    
    soup=BeautifulSoup(data,'lxml')
    tags=soup.table
    tag=tags.find_all('tr')
    x=[]
    for t in tag:
        x.append(t.get_text())
    
    for l in x:
        list1=l.split('\n')
        list2=list1[1:len(list1)-1]
        y.append(list2)
        
print(y[1:-3])

print(len(y))

# import pandas as pd
# df=pd.DataFrame(y)
# df
```

```python
final_data=[]
for j in y:
    if j[0]!="":
        final_data.append(j)
print(final_data)
```

```python
import csv
with open("17nepse.csv","w",newline="") as csvfile:
    wr=csv.writer(csvfile)
    wr.writerows(final_data)
```
