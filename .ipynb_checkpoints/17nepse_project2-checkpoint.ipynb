```python
# web scrpping of nepse

from bs4 import BeautifulSoup
import requests

n=int(input('enter n = '))
for i in range(1,n+1):
    url=BeautifulSoup(f'http://www.nepalstock.com/main/todays_price/index/{i}/','html.parser')
    req=requests.get(url)

req
```

```python
soup=req.text
soup
```

```python
soup=BeautifulSoup(soup,'lxml')
soup
```

```python
table=soup.table
table
```

```python
tabler=table.find_all('tr')
tabler
```

```python
data = []
for i in tabler:
    x=i.text
    x=x.split('\n')
    data.append(x[1:-1])
data
```

```python
final_data=[]
for j in data:
    if j[0]!="":
        final_data.append(j)
final_data
```

```python
import pandas as pd
df=pd.DataFrame(final_data)
dff=df.iloc[0:21]
dff
```

```python
dff.to_csv('nepsedata.csv')
dff
```

```python
# web scrpping of nepse

from bs4 import BeautifulSoup
import requests

n=int(input('enter n = '))
for i in range(1,n+1):
    url=BeautifulSoup(f'http://www.nepalstock.com/main/todays_price/index/{i}/','html.parser')
    req=requests.get(url)

    soup=req.text

    soupb=BeautifulSoup(soup,'lxml')

    table=soupb.table

    tabler=table.find_all('tr')

    data = []
    for i in tabler:
        x=i.text
        x=x.split('\n')
        data.append(x[1:-1])
    

    final_data=[]
    for j in data:
        if j[0]!="":
            final_data.append(j)
    

    import pandas as pd
    df=pd.DataFrame(final_data)
    dff=df.iloc[0:21]


    dff.to_csv('nepsedata.csv')
    dff
```
