```python
# web scrapping of corona virus 

from bs4 import BeautifulSoup
import requests                      # url use garne vayara requests vanne import gareko ho

url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup=requests.get(url)
soup
```

```python
soup=soup.text
```

```python
soup=BeautifulSoup(soup,'lxml')
soup
```

```python
table=soup.table
table
```

```python
tags=table.find_all('tr')
tags
```

```python
data=[]
for x in tags:
    data.append(x.text)
print(data)
```

```python
data=[]
for tag in tags:
    x=tag.text
    x=x.split("\n")
    data.append(x)
print(data)
```

```python
data=[]                
for tag in tags:
    x=tag.text
    x=x.split("\n")
    data.append(x[1:-1])
print(data)
```

```python

import pandas as pd
df=pd.DataFrame(data)
df
```

```python

df.to_csv('newww11111.csv')
df
```
