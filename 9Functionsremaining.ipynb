```python
# function with arguments and return type
# calculator

def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b

def cal(a,b,o):
    if o=='+':
        print(add(a,b))
    elif o=='-':
        print(sub(a,b))
    elif o=='*':
        print(mul(a,b))
    elif o=='/':
        if b!=0:
            print(div(a,b))
        else:
            print('the value of b cannot be zero.')
    else:
        print('invalid')

a=int(input('enter first number = '))
b=int(input('enter second number = '))
o=input('enter + - * / = ')

cal(a,b,o)
```
