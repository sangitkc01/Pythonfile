```python
# web scrapping of nepse

from bs4 import BeautifulSoup
import requests

url=BeautifulSoup('http://www.nepalstock.com/todaysprice','html.parser')
soup=requests.get(url)
soup
```

```python
soup.text
```

```python
soup=soup.text
```

```python
soup=BeautifulSoup(soup,'lxml')
soup
```

```python
table=soup.table
table
```

```python
table.find_all('tr')
```

```python
tags=table.find_all('tr')
tags
```

```python
data=[]
for tag in tags:
    x=tag.text
    x=x.replace('\n',' ')
    data.append(x)
print(data[0])
```

```python
data=[]
for tag in tags:
    x=tag.text
    x=x.split('\n')
    data.append(x)
print(data[0])
```

```python
data=[]
for tag in tags:
    x=tag.text
    x=x.split('\n')
    data.append(x[1:-1])
print(data)
```

```python
final_data=[]
for j in data:
    if j[0]!="":
        final_data.append(j)

print(final_data)
    
```

```python
import pandas as pd

df=pd.DataFrame(final_data)
df
```

```python
df=df.iloc[0:21]
df
```

```python
df=df.to_csv('webscrape_practical1_nepse.csv')
df
```
